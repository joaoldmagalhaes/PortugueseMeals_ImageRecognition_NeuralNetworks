```python
import urllib.request
from serpapi import GoogleSearch
import os

def serpapi_get_google_images(query, number_images):
    params = {
        "q": query,
        "tbm": "isch", #type image
        "ijn": "0", #page number
        'safe':'off', #no adult content
        'filter':'0', #no similar content
        "num": number_images, #number of photos to be scraped
        "api_key": "" #api key from serpapi
        }

    search = GoogleSearch(params) 
    results = search.get_dict() #transforming into dict
    images_results = results["images_results"] 
    
    # Download the images
    
    newpath = f'SerpApi_images/{query}'
    if not os.path.exists(newpath):
        os.makedirs(newpath)

    for index, image in enumerate(images_results, start=1):
        if index <= number_images: 
            print(f"Downloading {index} image...")
        
            opener=urllib.request.build_opener() #open the page
            opener.addheaders=[("User-Agent","Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.5060.114 Safari/537.36")]
            urllib.request.install_opener(opener) 

            urllib.request.urlretrieve(image["thumbnail"], f"SerpApi_images/{query}/thumbnail_size_img_{index}.jpg") #save images as thumbnailsize -> smaller than original
        else: 
            break
    
    print('Finished')
```
